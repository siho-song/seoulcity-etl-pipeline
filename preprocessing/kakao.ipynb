```python
import requests
import json
KAKAO_REST_API_KEY = '5f63993bd5eb200dfd4a0fda8879ca78' # 발급받은 API KEY
def convert_address_to_coordinates(address,count):
    """
    입력받은 주소를 WGS84 좌표계 좌표로 변환
    """

    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    header = {'Authorization': 'KakaoAK ' + KAKAO_REST_API_KEY}
 
    r = requests.get(url, headers=header)
    with open('kakao2.json','w') as f:
        json.dump(r.json(),f)
    print(count)
    if r.status_code == 200 and len(r.json()['documents']) != 0 and r.json() is not None :
        lng = float(r.json()['documents'][0]['road_address']['x'])
        lat = float(r.json()['documents'][0]['road_address']['y'])
    else:
        return 0.0,0.0
    
    return lat, lng

x_y_list=[]
count =0
for i in range(0,len(road_address)+1):
    x_y_list.append(convert_address_to_coordinates(road_address[i],count)) # 강남역 주소 
    count +=1

# Output
# (37.4981646510326, 127.028307900881)
```
